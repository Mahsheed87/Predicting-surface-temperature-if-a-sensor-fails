```python
df = df.drop('Timestamp',axis=1)
```

```python
X = df.drop('Surface_temperature_S321',axis=1).values
y = df['Surface_temperature_S321'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
df.shape
```

```python
model = Sequential()

model.add(Dense(18,activation='tanh'))
model.add(Dense(18,activation='tanh'))
model.add(Dense(18,activation='tanh'))
model.add(Dense(18,activation='tanh'))

model.add(Dense(1))

model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,validation_data=(X_test,y_test),batch_size=12,epochs=200)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
plt.ylabel('Loss',fontsize = 12, y = 0.5)
plt.xlabel('Epochs',fontsize = 12, y =0.5)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test,predictions)
```

```python
np.sqrt(mean_squared_error(y_test,predictions))
```

```python
mean_absolute_error(y_test,predictions)
```

```python
df['Surface_temperature_S321'].describe()
```

```python
explained_variance_score(y_test,predictions)
```

```python
plt.figure()
plt.scatter(y_test,predictions)
plt.plot(y_test,y_test,'r')
plt.ylabel('Prediction',fontsize = 12, y = 0.5)
plt.xlabel('True Values',fontsize = 12, y = 0.5)
```
